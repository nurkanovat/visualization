```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Load the CSV file into a DataFrame
file_path = r'C:\Users\kiril\visualization\02_activities\assignments\ssha-social-housing-summary-data.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the data structure
print(df.head())
```

```python
# Set the style for the plots
sns.set_style("whitegrid")

# Create a long-form DataFrame suitable for Seaborn
df_long = df.melt('Year', var_name='Quarter', value_name='Waiting List')

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Waiting List', hue='Quarter', data=df_long, marker='o')

# Customize the plot
plt.title('Social Housing Waiting List Trends by Quarter')
plt.xlabel('Year')
plt.ylabel('Number of People on Waiting List')
plt.xticks(df['Year'], rotation=45)
plt.legend(title='Quarter')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter data for the year 2014
df_2014 = df[df['Year'] == 2014]

# Display the 2014 data
print(df_2014)
```

```python
# Set the style for the plots
sns.set_style("whitegrid")

# Prepare the data for plotting
# Melt the DataFrame to long format
df_2014_long = df_2014.melt(id_vars=['Year'], var_name='Quarter', value_name='Waiting List')

# Create the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Quarter', y='Waiting List', data=df_2014_long, palette='viridis')

# Customize the plot
plt.title('Social Housing Waiting List by Quarter in 2014')
plt.xlabel('Quarter')
plt.ylabel('Number of People on Waiting List')
plt.ylim(85000, df_2014_long['Waiting List'].max() + 5000)  # Set y-axis starting point to 85000

# Show the plot
plt.tight_layout()
plt.show()
```
